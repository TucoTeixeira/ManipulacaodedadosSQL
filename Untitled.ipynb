Importando o arquivo e criando uma conexão com o mesmo.

```python
%sql sqlite:///salarydata.db
```

```python
#Criando tabela no banco de dados para receber o arquivo importado
%%sql

CREATE TABLE salary_data (
    country_name varchar(255) PRIMARY KEY,
    continent_name varchar(255),
    sage_spam varchar(255),
    median_salary decimal,
    average_salary decimal,
    lowest_salary decimal,
    highest_salary decimal
);
```

```python
data = pd.read_csv('salary_data.csv')
```

```python
data
```

```python
conn = sqlite3.connect('salarydata.db')
```

```python
data.to_sql('salary_data', conn, if_exists='replace', index=False)
```

```python
%%sql

SELECT * FROM salary_data
```

Agora que já importamos os arquivos estamos prontos para manipular os dados.

Qual seria a média de salário por continente? e qual deles possui a maior média? e a menor?

```python
%%sql

SELECT continent_name, CAST(AVG(average_salary) AS INTEGER) AS media_salario_inteiro
FROM salary_data
GROUP BY continent_name;
```

A partir dessa consulta podemos observar que a África é o continente com menor media salarial e a america do norte a que possui maior o maior valor.

Qual o pais com o menor valor de salário registrado, e o com maior?

```python
%%sql

SELECT country_name, lowest_salary
FROM salary_data
ORDER BY lowest_salary
LIMIT 1;
```

```python
%%sql

SELECT country_name, highest_salary
FROM salary_data
ORDER BY highest_salary DESC
LIMIT 1;
```

De acordo com as consultas acima apresentadas temos a Zambia como pais com menor salário não chegando sequer em 10 centavos de dólar, e a Suiça ocupando o pódio de salário registrados com 50.363,93 dólares.

Qual o pais da africa possui a maior mediana de salário?

```python
%%sql
SELECT country_name, median_salary
FROM salary_data
WHERE continent_name = 'Africa'
ORDER BY median_salary DESC
LIMIT 1;
```

qual pais da europa com menor valor médio de salário?

```python
%%sql
SELECT country_name, average_salary
FROM salary_data
WHERE continent_name = 'Europe'
ORDER BY average_salary
LIMIT 1;
```

qual a média mundial de salários?

```python
%%sql
SELECT AVG(average_salary) AS media_salario_total
FROM salary_data;
```
